###Model Architecture and Training Strategy

####1. An appropriate model architecture has been employed

My model consists of a convolution neural network with 5x5 & 3x3 filter sizes with depths between 24 and 64 (model.py, under Create Regression Model) The model includes RELU layers to introduce nonlinearity. The data was normalized prior to model implementation using a Keras lambda layer.

The model was trained and validated on different data sets to ensure that the model was not overfitting. The model was tested by running it through the simulator and ensuring that the vehicle could stay on the track.  I also validated the model output with a mean square loss per epoch plot.

###Model Architecture and Training Strategy

####1. Solution Design Approach

My first step was to use the NVIDIA convolution neural network model to train the car to steer. I thought this model might be appropriate because it had already been demonstrated to operate the car steering by the NVIDIA team.

In order to gauge how well the model was working, I split my image and steering angle data into a training and validation set. I found that my first model had a low mean squared error on the training set but a high mean squared error on the validation set.

To combat the overfitting, I modified the model so that it wold be trained on few epochs.

The final step was to run the simulator to see how well the car was driving around track one. There were a few areas where the vehicle drove up onto the edge of the track and then returned to the center of the track.  I tried the model several times and from time to time the car would drive off of the track and get stuck on the edge of the track unable to return to the cetner.  In order to improve the driving behavior  I added the training data supplied by the Udacity course to my training data to improve the driving behavior. I knew that the car was driving very well to begin with, but it needed more training data in order to understand what to do at specific points along the track.  Mainly the issue that need to be resolved was when the pavement for the track changed from one type to the next.  I knew that adding more training data might help the model to understand what to do when driving over pavement that changes color.

At the end of the process, the vehicle was able to drive autonomously around the track without leaving the road.

####2. Final Model Architecture

The NVIDIA architecture that I ended up implementing for my behavioral cloning neural network has an initial input layer that is then normalized via a lambda layer.  The lambda layer takes each pixel individually and divides that value by the maximum value of 255, so we will get a value range from 0 to 1 now instead of 0 to 255.  THen we take the value and mean center it about zero by subtracting 0.5, so not instead of a range of 0 to 1 we now have a range of -0.5 to 0.5.
After normalization comes the image cropping, which is done with the command Copping2D.  THe image is cropped 70 pixels down from the top and 25 pixels up from the bottom.  Next comes all five of the convolutional layers.  The first convolutional layer has a kernel of 5x5 and is filtered into 24 filters.  The next convolution has a kernel of 5x5 and is filtered into 36 filters. That is followed by a convolution with another 5x5 kernel and a filter output of 48.  It should be noted
that the first three convolutional layers also have relu activation functions associated with their outputs and the convolution does a 2x2 stride as well. The fourth convolution now has a kernel of 3x3 and filter size of 64.  Convolution number five also has a kernel of 3x3 and a filter size of 64.  The model data is then flattened into a single  data vector for use in the Dense functions.  These dense functions take the single data vector and condense the data down into a smaller vector.  
In this case the data is condensed down into a length of 100 and then down again into a length of 50, and then the data is condensed down into a final shape of 1x1 or a single data value.  In this case the single data value is the steering angle value needed to maneuver the car on the track.
